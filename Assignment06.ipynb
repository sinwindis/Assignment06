Importing Necessary Libraries

```python
import numpy as np
from numpy.linalg import inv
import matplotlib.pyplot as plt
```

provided code
generating data and ploting data

```python
num     = 201
std     = 20
a       = 2 
b       = 10

n       = np.random.rand(num)
nn      = n - np.mean(n)
x       = np.linspace(-100,100,num)
y1      = a * x + nn * std + b
y2      = a * x + b

plt.plot(x, y1, 'b.', x, y2, 'k--')
plt.show()

# x  : x-coordinate data
# y1 : (noisy) y-coordinate data
# y2 : (clean) y-coordinate data 
# y = f(x) = a * x + b
```

Generating Matrix
$$AX = B$$
$$A = [x, 1], X = [a, b], B = y1$$
$$\hat{X} = (A^TA)^{-1}A^TB$$

```python
A = np.empty((num, 2))
        
for i in range(num):
    A[i][0] = x[i]
    A[i][1] = 1

Xhat = np.matmul(np.matmul(inv(np.matmul(np.transpose(A), A)), np.transpose(A)), y1)



```

Ploting Least Square Approximation Graph

```python
ahat = Xhat[0]
stdhat = Xhat[1]
yhat = ahat * x + bhat

plt.plot(x, yhat, 'b-')
plt.show()
```
